## (おまけ)　ChEMBLのデータを描画

ChEMBLには構造情報としてSMILESが表示されるのでそれをコピーアンドペーストしても構造の描画ができます。

- [From ChEMBL](https://www.ebi.ac.uk/chembl/compound/inspect/CHEMBL941)

```python
glvc = Chem.MolFromSmiles("CN1CCN(Cc2ccc(cc2)C(=O)Nc3ccc(C)c(Nc4nccc(n4)c5cccnc5)c3)CC1")
```

```python
glvc
```

## sdfを扱う

```python
mols = Chem.SDMolSupplier("ch05_compounds.sdf")
```

```python
len(mols)
```

## 複数の分子をまとめて描画する

```python
Draw.MolsToGridImage(mols)
```

## (おまけ) for loopを回しながら一つずつ描画
